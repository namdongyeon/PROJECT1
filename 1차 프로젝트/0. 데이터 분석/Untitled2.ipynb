```python
import pandas as pd

master_set = pd.read_csv('C:\hmkd1\m4_db\dataset/마스터셋.csv', encoding='cp949', index_col=0)
master_set
```

```python
import pandas as pd

# 등급을 매길 컬럼들의 이름 리스트
columns = ['의류1Q', '의류2Q', '의류3Q', '의류4Q', '의류5Q', '의류6Q', '의류7Q', '의류8Q']

# 등급을 저장할 새로운 컬럼들의 이름 리스트
rank_columns = ['의류1Q_Rank', '의류2Q_Rank', '의류3Q_Rank', '의류4Q_Rank', '의류5Q_Rank', '의류6Q_Rank', '의류7Q_Rank', '의류8Q_Rank']

# 결과를 저장할 데이터프레임 생성
result_df = pd.DataFrame()

# 등급을 매기는 반복문
for i in range(len(columns)):
    column = columns[i]
    rank_column = rank_columns[i]
    
    # 데이터프레임에서 해당 컬럼을 기준으로 내림차순 정렬
    sorted_df = master_set.sort_values(by=column, ascending=False)
    
    # 해당 컬럼을 기준으로 순위 매기기
    sorted_df[rank_column] = pd.qcut(sorted_df[column][sorted_df[column] != 0], q=10, labels=False, duplicates='drop') + 1
    
    # 0인 값을 rank 1로 할당
    sorted_df.loc[sorted_df[column] == 0, rank_column] = 1
    
    # 결과를 결과 데이터프레임에 추가
    result_df = pd.concat([result_df, sorted_df[rank_column]], axis=1)

# 결과 확인
print(result_df)

result_df.to_excel('fe_가구.xlsx')
```
