```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df=pd.read_csv('./dataset/INCDECYEAR.csv', encoding='cp949')
df.head(3)
df
```

```python
df_kind=df.groupby('그룹').sum()
df_kind


df_kind['전체총액'].plot(kind='pie',
                     figsize=(8,6),
                     autopct='%.2f%%',
                     startangle=0)
plt.title('증가 감소 총액', size=20)
plt.axis('equal')
plt.legend(labels=df_kind.index, loc='best')
```

```python
df_kind=df.groupby('그룹').sum()
df_kind


df_kind['인원수'].plot(kind='pie',
                     figsize=(8,6),
                     autopct='%.2f%%',
                     startangle=0)
plt.title('증가감소 인원수 비율', size=20)
plt.axis('equal')
plt.legend(labels=df_kind.index, loc='best')
```

```python
df['14년도총구매금액'].plot(kind='bar')

plt.title('14년도총구매금액')
plt.xlabel('그룹')
plt.ylabel('총액')
```
