```python
# 분기별 구개금액이 0인 고객 제거한 master_set 불러오기

import warnings
warnings.filterwarnings('ignore')

import pandas as pd

master_set = pd.read_excel('/content/drive/MyDrive/예스리조(임시)/피처_엔지니어링/master_set.xlsx', index_col=0)
master_set.head()
```

```python
import pandas as pd

# 모든 열을 담을 리스트 생성
column_list = []

# master_set의 모든 열을 리스트에 추가
for col in master_set.columns:
    column_list.append(col)

# 결과 출력
print(column_list)
```

```python
import pandas as pd
import re

# 등급을 매길 컬럼들의 이름 리스트
columns = [col for col in master_set.columns if re.match(r".*[1-8]Q$", col)]

# 등급을 저장할 새로운 컬럼들의 이름 리스트
rank_columns = [col + "_Rank" for col in columns]

# 결과를 저장할 데이터프레임 생성
feature_df = pd.DataFrame()

# 등급을 매기는 반복문
for i in range(len(columns)):
    column = columns[i]
    rank_column = rank_columns[i]

    # 데이터프레임에서 해당 컬럼을 기준으로 내림차순 정렬
    sorted_df = master_set.sort_values(by=column, ascending=False)

    # 해당 컬럼을 기준으로 순위 매기기
    sorted_df[rank_column] = pd.qcut(sorted_df[column][sorted_df[column] != 0], q=10, labels=False, duplicates='drop') + 1

    # 0인 값을 rank 1로 할당
    sorted_df.loc[sorted_df[column] == 0, rank_column] = 1

    # 결과를 결과 데이터프레임에 추가
    feature_df = pd.concat([feature_df, sorted_df[rank_column]], axis=1)

# 결과 확인
feature_df.head()

# result_df.to_csv('등급결과.csv')
# result_df.to_excel('feature_df.xlsx')
```

```python
import pandas as pd

# 원본 데이터프레임에서 원하는 컬럼들을 선택하여 새로운 데이터프레임 생성
feature_df_5q = feature_df.filter(regex=".*[1-5]Q_Rank$", axis=1)

# 새로운 데이터프레임 출력
feature_df_5q
```

```python
import pandas as pd

# 새로운 데이터프레임 생성
new_df = pd.DataFrame()

# 원하는 계산을 수행하여 새로운 컬럼 추가
categories = ['A', 'B', 'C', 'D', '가공', '신선', '외식/편의시설', '일상', '가구', '스포츠', '의류', '패션잡화',\
              '유아/아동', '명품', '가전', '교육/문화', '뷰티', '의료', 'TOTAL', 'VIS']  # 카테고리 리스트

for category in categories:
    category_cols = [col for col in feature_df_5q.columns if col.startswith(category)]

    for i in range(len(category_cols) - 1):
        curr_col = category_cols[i]
        next_col = category_cols[i + 1]
        new_col = next_col + ' - ' + curr_col
        new_df[new_col] = feature_df_5q[next_col] - feature_df_5q[curr_col]

# 새로운 데이터프레임 출력
new_df
```

```python
import pandas as pd

# 새로운 데이터프레임 생성
result_df = pd.DataFrame()

# 각 카테고리별로 합계 및 절댓값 합계 계산
categories = ['A', 'B', 'C', 'D', '가공', '신선', '외식/편의시설', '일상', '가구', '스포츠', '의류', '패션잡화',\
              '유아/아동', '명품', '가전', '교육/문화', '뷰티', '의료', 'TOTAL', 'VIS']  # 카테고리 리스트

for category in categories:
    category_cols = [col for col in new_df.columns if col.startswith(category)]

    # 합계 계산
    sum_col = f'{category}_Sum'
    result_df[sum_col] = new_df[category_cols].sum(axis=1)

    # 절댓값 합계 계산
    abs_sum_col = f'{category}_Abs_Sum'
    result_df[abs_sum_col] = new_df[category_cols].abs().sum(axis=1)

# 새로운 데이터프레임 출력
result_df = result_df.sort_index()
result_df
```
